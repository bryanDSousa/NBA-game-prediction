# Limpieza Datos para Evaluar los Jugadores más Importantes por Equipo

Primero, importaremos las librerías que necesitaremos para el desarrollo del código.

```python
import pandas as pd
import numpy as np
import calendar
```

Seguidamente, importaremos el csv que utilizaremos en un futuro y declararemos las variables a usar: 
    stats_jugadores: variable utilizada para guardar el csv en formato dataframe.
    year_filtered: array utilizado para filtrar el dataframe inicial por las temporadas deseadas.
    dic_trx_mes_numero: diccionario utilizado para realizar la conversión de mes identifica con tres letras a mes identificado con un número.
    dic_df_jugador: diccionario utilizado para almacenar cada dataframe con las estadísticas de cada jugador.

```python
stats_jugadores = pd.read_csv("./Stats_jugadores_10_04_20.csv")
year_filtered = [2019, 2020]
dic_trx_mes_numero= {}
dic_df_jugador = {}
```

```python
pd.options.display.max_columns=60
pd.options.display.max_rows=999
stats_jugadores
```

Filtramos el dataset mediante year_filtered. Creamos las variables 
mes_dia: array cuyo contenido resulta de la separación de la columna Date en mes y día.
equipos_unicos: contiene todos los equipos del df
jugadores_unicos: contiene todos los jugadores del df
columnas_df: contiene todas las columnas del df
columnas_jugadores_df: contiene todas las columnas de estadísticas de los jugadores
Posteriormente damos formato de string al año y de int a las variables que nos pueden ser decimales.

```python
#filtrado dataset
stats_jugadores['Season'] = stats_jugadores['Season'].astype('int32')
stats_jugadores = stats_jugadores[stats_jugadores.Season.isin(year_filtered)]

#declaración de variables
mes_dia = stats_jugadores["Date"].str.split(" ", n = 1, expand = True) 
equipos_unicos = stats_jugadores['Team'].unique()
columnas_df = stats_jugadores.columns
columnas_jugadores_df = columnas_df[11:46]

#adaptación tipos variables
stats_jugadores['Year'] = stats_jugadores['Year'].astype('int32').astype('str')
stats_jugadores[['Puntos','Faltas','Perdidas','Robos', 'Asistencias','Rebotes total','Rebotes ofensivos','Tiros libres intentados', 'Tiros libres anotados','Tiros de tres intentados','Tiros de tres anotados','Tiros intentados','Team Points', 'Opponent Points','Tiros anotados',"Rebotes defensivo", "Tapones"]] = stats_jugadores[['Puntos','Faltas','Perdidas','Robos', 'Asistencias','Rebotes total','Rebotes ofensivos','Tiros libres intentados', 'Tiros libres anotados','Tiros de tres intentados','Tiros de tres anotados','Tiros intentados','Team Points', 'Opponent Points','Tiros anotados',"Rebotes defensivo", "Tapones"]].astype('int32')
```

Cambiamos el tipo de los minutos de string a float. Para ello hay que cambiar : por .

```python
stats_jugadores['Minutes Played'] = stats_jugadores['Minutes Played'].replace('\:', '.', regex = True)
stats_jugadores['Minutes Played'] = pd.to_numeric(stats_jugadores['Minutes Played'])
```

Limpiamos el ID para quedarnos con la información valiosa

```python
stats_jugadores['ID Partido'] = stats_jugadores['ID Partido'].replace('[a-z]', '', regex = True)
stats_jugadores['ID Partido'] = stats_jugadores['ID Partido'].replace('\.', '', regex = True)
stats_jugadores['ID Partido'] = stats_jugadores['ID Partido'].replace('\/', '', regex = True)
```

Creamos las columnas mes y día y eliminamos la columna Date

```python
# making separate first name column from new data frame 
stats_jugadores["Month"]= mes_dia[0] 
stats_jugadores['Month'] = stats_jugadores['Month'].astype('str')

# making separate last name column from new data frame 
stats_jugadores["Day"]= mes_dia[1] 
stats_jugadores['Day'] = stats_jugadores['Day'].astype('str') 

# Dropping old Name columns 
stats_jugadores.drop(columns =["Date"], inplace = True)
```

Creamos un diccionario para relacionar el mes con letra con el mes con su número.

```python
for k,v in enumerate(calendar.month_abbr):
    dic_trx_mes_numero[v] = k
```

Transformamos el mes con letra a mes con número y recalculamos la columna result ya que en el dataset anterior estaba mal calculada. Además creamos la columna Complete Date con la fecha de cada partido y por último elinamos las columnas Year, Day, Month

```python
for indice in stats_jugadores.index:
    if(stats_jugadores.loc[indice,'Month'] in dic_trx_mes_numero.keys()):
        stats_jugadores.loc[indice,'Month'] = dic_trx_mes_numero[stats_jugadores.loc[indice,'Month']]
    if (stats_jugadores.loc[indice,'Team Points'] > stats_jugadores.loc[indice,'Opponent Points']):
        stats_jugadores.loc[indice,'Result'] = 1
    else: 
        stats_jugadores.loc[indice,'Result'] = 0
stats_jugadores['Complete Date'] = stats_jugadores['Year'].astype('str') + stats_jugadores['Month'].astype('str') + stats_jugadores['Day'].astype('str')
stats_jugadores['Complete Date'] = pd.to_datetime(stats_jugadores['Complete Date'], format='%Y%m%d', errors='ignore')
stats_jugadores.drop(columns =["Year", "Day", "Month"], inplace = True) 
```

Recorremos cada equipo del dataset para obtener un dataframe por equipo. Una vez realizado esto obtenemos un dataframe por jugador con la variables de las estadísticas de los jugadores con el nombre de cada jugador.

```python
contador = 0
concatenado = '0'

for x in equipos_unicos:
    df1 = stats_jugadores[stats_jugadores['Team'] == x]
    jugadores_unicos = df1['Name'].unique()
    for y in jugadores_unicos:
        df2 = df1[df1['Name'] == y]
        for z in range(len(columnas_jugadores_df)):
            concatenado = y + '_' + columnas_jugadores_df[z]
            df2.rename(columns={columnas_jugadores_df[z] : concatenado}, inplace = True)
        dic_df_jugador[contador] = df2
        contador += 1       
    
```

Creamos un dataframe por cada equipo con granularidad partido y con cada estadística de cada jugador en columna.


```python
contador = 0
dic_final_equipos = {}
dic_final_equipos[0] = dic_df_jugador[0]
for i in range(1,len(dic_df_jugador)):
    if (dic_df_jugador[i].loc[dic_df_jugador[i].index[0], 'Team'] == equipos_unicos[contador]):
        dic_df_jugador[i] = dic_df_jugador[i].drop(columns =['Team', 'Season', "Local", 'Opponent', 'Team Points', 'Opponent Points', 'Result', 'Name', 'Complete Date'])
        dic_final_equipos[contador] = pd.merge(dic_final_equipos[contador], dic_df_jugador[i], on = 'ID Partido', how='outer')
    else:
        contador += 1
        dic_final_equipos[contador] = dic_df_jugador[i]
```
