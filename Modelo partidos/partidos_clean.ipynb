```python
data = data.dropna()
```

```python
# Columna Fecha

data['Year'] = data['Year'].astype('int32').astype('str')
new = data["Date"].str.split(" ", n = 1, expand = True) 
  
data["Month"]= new[0] 
  
data["Day"]= new[1] 
  
data.drop(columns =["Date"], inplace = True)
```

```python
# Columna Fecha
import calendar
mes_numero= {}
for k,v in enumerate(calendar.month_abbr):
    mes_numero[v] = k
```

```python
# Columna Fecha
data['Date'] = data['Year'].astype('str') + data['Month'].astype('str')+ data['Day'].astype('str')

data['Date'] = pd.to_datetime(data['Date'].astype(str), format='%Y%m%d', errors='ignore')

data.drop(columns =["Year", "Day", "Month"], inplace = True)
```

```python
variables = ['boxcore_url','Date', 'Season', #variables de interes del partido
             'local_team', 'visitor_team', 'local_pts', 'visitor_pts', 'Result', 
             #estadisticas del local
             'local_Lose','local_Win', 'local_Percentagewl', 'local_3pa_rate',
             'local_Conf_position',
             'local_Home_lose', 'local_Home_win',
             'local_ast', 'local_ast_pct','local_blk','local_blk_pct','local_def_rate','local_drb','local_drb_pct',
             'local_effective_fg_pct','local_fg','local_fg3','local_fg3_pct','local_fg3a','local_fg_pct','local_fga',
             'local_ft','local_ft_pct','local_fta','local_fta_rate','local_off_rate','local_orb','local_orb_pct',
             'local_pf','local_stl','local_stl_pct','local_tov','local_tov_pct',
             'local_trb_pct','local_trn', 'local_true_shooting_pct',
             #estadisticas del visitante
             'visitor_Lose','visitor_Win', 'visitor_Percentagewl', 'visitor_3pa_rate',
             'visitor_Conf_position',
             'visitor_Away_lose', 'visitor_Away_win',
             'visitor_ast', 'visitor_ast_pct','visitor_blk','visitor_blk_pct','visitor_def_rate','visitor_drb','visitor_drb_pct',
             'visitor_effective_fg_pct','visitor_fg','visitor_fg3','visitor_fg3_pct','visitor_fg3a','visitor_fg_pct','visitor_fga',
             'visitor_ft','visitor_ft_pct','visitor_fta','visitor_fta_rate','visitor_off_rate','visitor_orb','visitor_orb_pct',
             'visitor_pf','visitor_stl','visitor_stl_pct','visitor_tov','visitor_tov_pct',
             'visitor_trb_pct','visitor_trn', 'visitor_true_shooting_pct'
            ]
data_clean = data[variables]
data_clean.head(20)
```

```python
data_clean = data_clean.dropna()
len(data_clean)
```
